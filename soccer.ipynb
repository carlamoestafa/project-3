```python
query="""
SELECT match_api_id,home_team_goal, away_team_goal,
COUNT(DISTINCT match_api_id) AS tie_match
FROM Match
GROUP BY match_api_id
HAVING home_team_goal == away_team_goal
;
"""
tie_matches = pd.read_sql_query(query, conn)
tie_matches
```

```python
tie_matches.sum()
```

```python
query="""
SELECT SUM(tie_match) FROM
(SELECT match_api_id,home_team_goal, away_team_goal,
COUNT(DISTINCT match_api_id) AS tie_match
FROM Match
GROUP BY match_api_id
HAVING home_team_goal == away_team_goal) AS total_tie_matches;
"""
tie_matches = pd.read_sql_query(query, conn)
tie_matches
```
